```python
import numpy as np
import pandas as pd

# Exercise 8a 
college = pd.read_csv("../datasets/College.csv")
college.head()
```

```python
# Exercise 8b
college2 = pd.read_csv('../datasets/College.csv', index_col =0)
college3 = college.rename ({'Unnamed: 0': 'College'}, axis =1)
college3 = college3.set_index('College')
college = college3
```

```python
# Exercise 8c
college.describe()
```

```python
# Exercise 8d
import matplotlib.pyplot as plt
pd.plotting.scatter_matrix(college[['Top10perc', 'Apps', 'Enroll']])
plt.show()
```

```python
# Exercise 8e
college.boxplot(column='Outstate', by='Private', figsize=(8, 6))
plt.title('Out-of-State Tuition by Private/Public Colleges')
plt.suptitle('')
plt.xlabel('Private')
plt.ylabel('Outstate Tuition')

plt.show()
```

```python
# Exercise 8f
college['Elite'] = pd.cut(college['Top10perc'], [0, 50, 100],
                          labels=['No', 'Yes'])

print(college['Elite'].value_counts())

college.boxplot(column='Outstate', by='Elite', figsize=(8, 6))
plt.show()

```

```python
# Exercises 8g
# Choose the variables and number of bins
variables = ['Apps', 'Accept', 'Enroll', 'Outstate']
bins_list = [10, 20, 15, 30]  # Different bin sizes for each

# Create a 2x2 grid of subplots
fig, axes = plt.subplots(2, 2, figsize=(12, 10))

# Flatten the axes array for easy iteration
axes = axes.flatten()

# Loop through each variable and plot its histogram
for i, var in enumerate(variables):
    college[var].plot.hist(bins=bins_list[i], ax=axes[i], alpha=0.7)
    axes[i].set_title(f'Histogram of {var}')
    axes[i].set_xlabel(var)
    axes[i].set_ylabel('Frequency')

# Adjust layout
plt.tight_layout()
plt.show()
```
